# Modeling

```python
x = dataset.drop(columns=["price","model","transmission","fuelType"])
y = dataset["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
coef_dict = {
    'features' : x.columns,
    'coef_value' : lr.coef_
}
coef = pd.DataFrame(coef_dict, columns=['features', 'coef_value'])
coef
```

```python
lr.predict(x_test)
```
