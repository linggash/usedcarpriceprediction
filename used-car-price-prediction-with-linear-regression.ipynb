```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('../input/used-car-dataset-ford-and-mercedes/audi.csv')
dataset.head()
```

```python
dataset['engineSize'] = dataset['engineSize'].replace(0,1)
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(data = dataset, x_vars=['year', 'mileage', 'tax', 'mpg', 'engineSize'], y_vars=['price'], height=5, aspect=1)
```

```python
dataset.corr().style.background_gradient().set_precision(1)
```

# Modeling

```python
x = dataset.drop(columns=["price","model","transmission","fuelType"])
y = dataset["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
lr.coef_
```

# Prediksi dan Evaluasi Model

```python
pred_y = lr.predict(x_test)

print(pred_y[:5])
print(y_test[:5])
```

```python
ev = {
    'asli' : y_test[:5],
    'prediksi' : pred_y[:5]
}
tabel = pd.DataFrame(ev, columns=['asli', 'prediksi'])
tabel
```

```python
lr.score(x_test, y_test)
```
